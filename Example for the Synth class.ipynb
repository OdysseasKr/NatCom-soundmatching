# How to use the Synth class

The Synth class represents a synthesizer and is used to generate sounds given a set of parameters. The synthesizer contains two oscillators (the things that create the waves of different shapes) and a Filter that smoothens the hard edges. Here's how to use it.

```python
from synth import Synth
from utils import play_sound, plot_sound
mysynth = Synth()
```

First we need to set the parameters. The parameters are:
1. The shape of each oscillator
2. The amplitude of each oscillator (how loud it is)
3. The phase of each oscillator.
4. The cutoff of the filter (the lower cutoff, the smoother the wave becomes)

```python
mysynth.set_parameters(osc_1='SquareH', cutoff=2500)
```

Now that we have set the parameters, we can play notes and get the sound. The note and duration variables are optional.

Here we use `note=440` which means a note at 440Hz which is C.

```python
sound = mysynth.get_sound_array(note=440, duration=1)
print(sound)
play_sound(sound)
plot_sound(sound)
```

If you want to specify all parameters explicitely, you can do the following:

```python
params = {
    'osc_1': 'Sine', # Wave type for oscillator 1
    'osc_2': 'SquareH', # Wave type for oscillator 2
    'amp_1': 0.5, # Amplityde for oscillator 1
    'amp_2': 0.5, # Amplityde for oscillator 2
    'phase_1': 0, # Phase shift for oscillator 1
    'cutoff': 5000 # Cutoff for the Filter
}
mysynth.set_parameters(**params)
```

and again play and plot the sound as usual

```python
sound = mysynth.get_sound_array(note=440, duration=1)
print(sound)
play_sound(sound)
plot_sound(sound)
```
